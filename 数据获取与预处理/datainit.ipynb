```python
df_movies = pd.read_csv("../../数据集/豆瓣电影数据集/豆瓣电影数据集/movies.csv")
df_movies
```

```python
df_mv = df_movies.copy()

# 将指定列的数据以'/'为分隔符拆分并转化为列表
df_mv['ACTORS'] = df_movies['ACTORS'].str.split('/')
df_mv['DIRECTORS'] = df_movies['DIRECTORS'].str.split('/')
df_mv['GENRES'] = df_movies['GENRES'].str.split('/')
df_mv['LANGUAGES'] = df_movies['LANGUAGES'].str.split('/')
df_mv['REGIONS'] = df_movies['REGIONS'].str.split('/')

# 删除NAME列中结尾的" - 电影"
df_mv['NAME'] = df_mv['NAME'].str.replace(' - 电影$', '', regex=True)

# 删除六列字段中数据为空的行
df_mv = df_mv[['NAME', 'ACTORS', 'DIRECTORS', 'GENRES', 'LANGUAGES', 'REGIONS']].dropna()
df_mv.reset_index(drop=True, inplace=True) # 重新排序索引

# 由于部分类型以繁体字形式出现且有现成的类型与之对应，因此考虑统一格式
df_mv['GENRES'] = df_mv['GENRES'].apply(lambda x: ['剧情' if genre == '劇情 Drama' 
                                       else '惊悚' if genre == '驚悚 Thriller'
                                       else '悬疑' if genre in '懸疑 Mystery'
                                       else '真人秀' if genre == 'Reality-TV'
                                       else '喜剧' if genre in ['喜劇 Comedy', 'Comedy']
                                       else '传记' if genre == '傳記 Biography'
                                       else '历史' if genre == '歷史 History'
                                       else '动作' if genre == '動作 Action'
                                       else '成人' if genre == 'Adult'
                                       else genre for genre in x])

df_mv
```

```python
df_tmdb = pd.read_csv("../../数据集/TMDB数据集/tmdb_5000_credits.csv")
df_tmdb = df_tmdb[['title', 'cast', 'crew']]
df_tmdb
```

```python
df_tmdb_mv = pd.read_csv("../../数据集/TMDB数据集/tmdb_5000_movies.csv")
df_tmdb_mv = df_tmdb_mv[['title', 'genres', 'production_countries']]
df_tmdb_mv['language'] = '英语'
df_tmdb_mv
```

### 数据初步整理

```python
# 从JSON中提取数据
import json

# 解析函数

# 提取json中的name
def extract_name(actors_str):
    # 用于存储结果的列表
    results = []
    actors = json.loads(actors_str)  # 将字符串解析为 JSON 对象
    for actor in actors:
        results.append(actor['name'])  # 提取 name 
    return results

# 提取职员中的导演
def extract_director_names(data_str):
    directors = json.loads(data_str)  # 将字符串解析为 JSON 对象
    results = []  # 初始化结果列表
    
    # 遍历所有条目，提取 job 为 Director 的 name 和 gender
    for director in directors:
        if director['job'] == 'Director':
            results.append(director['name'])
    return results

df1 = df_tmdb['cast'].apply(extract_name)
df2 = df_tmdb['crew'].apply(extract_director_names)
df3 = df_tmdb_mv['genres'].apply(extract_name)
df4 = df_tmdb_mv['production_countries'].apply(extract_name)

df_t_mv = pd.concat([df_tmdb['title'], df1, df2, df3, df_tmdb_mv['language'], df4], axis=1) # 将两个表对应的字段重新组合
df_t_mv.columns = ['NAME', 'ACTORS', 'DIRECTORS', 'GENRES', 'LANGUAGES', 'REGIONS'] # 修改列名（与豆瓣电影集匹配）
df_t_mv = df_t_mv[df_t_mv.apply(lambda row: all(bool(cell) for cell in row), axis=1)] # 删除内容为空的数据行
df_t_mv.reset_index(drop=True, inplace=True) # 重新排序索引
df_t_mv
```

### 实体信息汇总

```python
# 制片国家/地区
df_t_regions = pd.DataFrame(df_t_mv['REGIONS'].explode('REGIONS'))
df_t_regions = df_t_regions.drop_duplicates()
df_t_regions = df_t_regions.dropna()
df_t_regions.reset_index(drop=True, inplace=True) # 重新排序索引
df_t_regions
```

```python
# 类型
df_t_genres = pd.DataFrame(df_t_mv['GENRES'].explode('GENRES'))
df_t_genres = df_t_genres.drop_duplicates()
df_t_genres = df_t_genres.dropna()
df_t_genres.reset_index(drop=True, inplace=True) # 重新排序索引
df_t_genres
```

```python
# 演员与导演

# 用于存储结果的列表
results_actor = []
results_director = []

# 提取演员的名字和性别
def extract_names_and_genders(actors_str):
    actors = json.loads(actors_str)  # 将字符串解析为 JSON 对象
    for actor in actors:
        results_actor.append({'name': actor['name'], 'gender': actor['gender']})  # 提取 name 和 gender

# 提取职员中的导演和性别
def extract_director_names_and_genders(data_str):
    directors = json.loads(data_str)  # 将字符串解析为 JSON 对象
    
    # 遍历所有条目，提取 job 为 Director 的 name 和 gender
    for director in directors:
        if director['job'] == 'Director':
            results_director.append({'name': director['name'], 'gender': director['gender']})

# 应用解析函数并添加新的列
df_tmdb['cast'].apply(extract_names_and_genders)
df_tmdb['crew'].apply(extract_director_names_and_genders)
# 创建新的 DataFrame
df_tmdb_actor = pd.DataFrame(results_actor)
df_tmdb_director = pd.DataFrame(results_director)

df_tmdb_actor = df_tmdb_actor.drop_duplicates()
df_tmdb_actor.reset_index(drop=True, inplace=True)
df_tmdb_director = df_tmdb_director.drop_duplicates()
df_tmdb_director.reset_index(drop=True, inplace=True)
```

```python
import ast
df_all_mv = pd.read_csv("../../数据集/处理后的数据集/movie_data_C3.csv")

# 将存储的字符串值修改成列表形式
df_c = df_all_mv[['ACTORS','DIRECTORS','GENRES','LANGUAGES','REGIONS']].copy()

# 将列中的字符串转换为列表
df_all_mv['ACTORS'] = df_c['ACTORS'].apply(ast.literal_eval)
df_all_mv['DIRECTORS'] = df_c['DIRECTORS'].apply(ast.literal_eval)
df_all_mv['GENRES'] = df_c['GENRES'].apply(ast.literal_eval)
df_all_mv['REGIONS'] = df_c['REGIONS'].apply(ast.literal_eval)

# 处理 LANGUAGES 列
def convert_languages(lang):
    if isinstance(lang, str) and lang == "英语":  # 直接字符串"英语"
        return [lang]  # 转换为列表
    else:
        return ast.literal_eval(lang)  # 否则正常转换
df_all_mv['LANGUAGES'] = df_c['LANGUAGES'].apply(convert_languages)

df_all_mv
```

### 实体信息统计

```python
# 构建统计各字段实体信息数据的函数
def count_entity(df, str_name):
    df_tt = pd.DataFrame(df[str_name].explode(str_name))
    df_tt.replace('\s','',regex=True,inplace=True)
    df_tt = df_tt.drop_duplicates()
    df_tt = df_tt.dropna()
    df_tt.reset_index(drop=True, inplace=True) # 重新排序索引
    return df_tt

dataframes = {
    'ACTORS': count_entity(df_all_mv,'ACTORS'),
    'DIRECTORS':count_entity(df_all_mv, 'DIRECTORS'), 
    'GENRES': count_entity(df_all_mv, 'GENRES'),
    'LANGUAGES': count_entity(df_all_mv, 'LANGUAGES'),
    'REGIONS': count_entity(df_all_mv, 'REGIONS')
}

# 统计每个 DataFrame 的行数
data_stats = {
    name: df.shape[0] for name, df in dataframes.items()
}
data_stats['MOVIES'] = df_all_mv['NAME'].count()

# 创建新的 DataFrame
df_count = pd.DataFrame(data_stats.items(), columns=['实体信息', '数据量'])
df_count
```
