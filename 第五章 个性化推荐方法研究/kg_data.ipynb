```python
import ast

df_all = pd.read_csv("../../数据集/处理后的数据集/user_data_2_C4.csv")
# 将存储的字符串值修改成列表形式
df_c = df_all[['ACTORS','DIRECTORS','GENRES','LANGUAGES','REGIONS']].copy()

# 将列中的字符串转换为列表，处理NaN值
df_all['ACTORS'] = df_c['ACTORS'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
df_all['DIRECTORS'] = df_c['DIRECTORS'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
df_all['GENRES'] = df_c['GENRES'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
df_all['REGIONS'] = df_c['REGIONS'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
df_all['LANGUAGES'] = df_c['LANGUAGES'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])

df_all
```

```python
df_filter = pd.read_csv("../../数据集/处理后的数据集/recommendation_C5.csv")
df_filter
```

```python
unique_movie_ids = df_filter['MOVIE_ID'].unique()

# 根据去重后的 MOVIE_ID 从 df_all 提取 GENRES 和 ACTORS
filtered_data = df_all[df_all['MOVIE_ID'].isin(unique_movie_ids)][['MOVIE_ID', 'NAME', 'GENRES', 'ACTORS', 'DIRECTORS']]
# 对 filtered_data 进行去重，保留第一个出现的 MOVIE_ID
filtered_data = filtered_data.drop_duplicates(subset='MOVIE_ID', keep='first')

# 自定义函数删除“剧情”
def remove_drama(genres):
    return [genre for genre in genres if genre != '剧情']

# 应用函数到 GENRES 列
filtered_data['GENRES'] = filtered_data['GENRES'].apply(remove_drama)
filtered_data['GENRES'] = filtered_data['GENRES'].apply(lambda x: x[:3])
filtered_data['GENRES'] = filtered_data['GENRES'].apply(lambda x: sorted(x))
filtered_data['ACTORS'] = filtered_data['ACTORS'].apply(lambda x: x[:3])
filtered_data['DIRECTORS'] = filtered_data['DIRECTORS'].apply(lambda x: x[:3])
filtered_data = filtered_data.reset_index()
filtered_data
```

```python
# filtered_data['GENRES'] = filtered_data['GENRES'].apply(lambda x: x[:3])
# 排序后再转为 tuple 去重
unique_combinations = filtered_data['GENRES'].apply(lambda x: tuple(sorted(x))).drop_duplicates()

# 转回列表并打印
unique_combinations_list = unique_combinations.tolist()
print(f"共有 {len(unique_combinations_list)} 种不同的组合（顺序无关）：")
for combo in unique_combinations_list:
    print(list(combo))
```

```python
exploded_data = filtered_data.explode('GENRES')
exploded_data = exploded_data.explode('ACTORS')
exploded_data = exploded_data.explode('DIRECTORS')
exploded_data = exploded_data.drop(['NAME', 'index'], axis=1)
# 2. 对 GENRES 列唯一值编码为 G1, G2, ...
unique_genres = exploded_data['GENRES'].dropna().unique()
genre_mapping = {genre: f'G{i+1}' for i, genre in enumerate(unique_genres)}

# 3. 对 ACTORS 列唯一值编码为 A1, A2, ...
unique_actors = exploded_data['ACTORS'].dropna().unique()
actor_mapping = {actor: f'A{i+1}' for i, actor in enumerate(unique_actors)}

# 4. 对 DIRECTORS 列唯一值编码为 A1, A2, ...
unique_directors = exploded_data['DIRECTORS'].dropna().unique()
director_mapping = {director: f'D{i+1}' for i, director in enumerate(unique_directors)}

# 5. 应用编码映射
exploded_data['GENRE_CODE'] = exploded_data['GENRES'].map(genre_mapping)
exploded_data['ACTOR_CODE'] = exploded_data['ACTORS'].map(actor_mapping)
exploded_data['DIRECTORS_CODE'] = exploded_data['DIRECTORS'].map(director_mapping)
exploded_data
```

```python
df_kgmv = exploded_data.copy()
df_kgmv = df_kgmv[['MOVIE_ID', 'GENRE_CODE', 'ACTOR_CODE', 'DIRECTORS_CODE']]
df_kgmv
```

```python
# 1. 将四列合并为一个 Series
all_entities = pd.concat([
    df_kgmv['MOVIE_ID'],
    df_kgmv['GENRE_CODE'],
    df_kgmv['ACTOR_CODE'],
    df_kgmv['DIRECTORS_CODE']
])

# 2. 去重 & 重置索引
unique_entities = all_entities.dropna().unique()  # 去掉可能的 NaN
entity2id = {entity: idx for idx, entity in enumerate(unique_entities)}

# 3. 生成一个新的 DataFrame
entity_df = pd.DataFrame({
    'entity': list(entity2id.keys()),
    'id': list(entity2id.values())
})

# 4. 输出为 entity2id.txt，使用制表符分隔
entity_df.to_csv('../../数据集/处理后的数据集/三元组/entity2id.txt', sep='\t', index=False, header=False)
```

```python
# 1. 为每种关系构造三元组 DataFrame
triplets_genre = df_kgmv[['MOVIE_ID', 'GENRE_CODE']].dropna().copy()
triplets_genre['relation'] = 'Relation_m_g'
triplets_genre.rename(columns={'GENRE_CODE': 'tail'}, inplace=True)

triplets_actor = df_kgmv[['MOVIE_ID', 'ACTOR_CODE']].dropna().copy()
triplets_actor['relation'] = 'Relation_m_a'
triplets_actor.rename(columns={'ACTOR_CODE': 'tail'}, inplace=True)

triplets_director = df_kgmv[['MOVIE_ID', 'DIRECTORS_CODE']].dropna().copy()
triplets_director['relation'] = 'Relation_m_d'
triplets_director.rename(columns={'DIRECTORS_CODE': 'tail'}, inplace=True)

# 2. 合并三种三元组
triplets = pd.concat([triplets_genre, triplets_actor, triplets_director], ignore_index=True)
triplets.rename(columns={'MOVIE_ID': 'head'}, inplace=True)

# 3. 重新排序列：头实体, 关系, 尾实体
triplets = triplets[['head', 'relation', 'tail']]

# 4. 保存为 train.txt，制表符分隔，无列名
triplets.to_csv('../../数据集/处理后的数据集/三元组/train.txt', sep='\t', index=False, header=False)
```

# 处理结果（计算过程在云服务器）

```python
import numpy as np
import pandas as pd
from scipy.spatial.distance import pdist, squareform

# 假设你的文件名是 'data.txt'
filename = 'torch_TransE_entity_50dim_batch5'

data = pd.read_csv(filename, sep='\t', header=None, names=['index', 'vector'])

# 将 'vector' 列中的字符串转换为 NumPy 数组
data['vector'] = data['vector'].apply(lambda x: np.array(eval(x)))

# 将所有向量放在一个 NumPy 数组中
vectors = np.stack(data['vector'].values)

# 计算欧几里得距离矩阵
distance_matrix = squareform(pdist(vectors, metric='euclidean'))

# 创建一个新的矩阵来存储结果
result_matrix = np.zeros_like(distance_matrix)

# 对角线元素设为 1，非对角线元素 +1 后取倒数
for i in range(len(distance_matrix)):
    for j in range(len(distance_matrix)):
        if i == j:  # 对角线元素
            result_matrix[i, j] = 1
        else:  # 非对角线元素
            result_matrix[i, j] = 1 / (distance_matrix[i, j] + 1)

# 将结果矩阵转为 DataFrame 方便查看
result_df = pd.DataFrame(result_matrix)

result_df
```
