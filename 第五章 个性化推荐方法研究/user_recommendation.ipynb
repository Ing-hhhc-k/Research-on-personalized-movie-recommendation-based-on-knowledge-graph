```python
'''
基于用户的推荐算法
'''
from math import sqrt,pow
import operator
from scipy.stats import pearsonr

class UserCf():

    #获得初始化数据,计算每个用户的评分的平均值
    def __init__(self,data):
        self.data=data
        self.ave = {}
        self.max = 0
        for key,value in self.data.items():
            sum1 = 0.0
            for item,score in value.items():
                if(int(item)>self.max):
                    self.max = int(item)
                sum1 += score
            self.ave[key] = sum1/len(data[key])

    #1.获取用户待预测分数的相关用户，如找到用户C中待评分商品4的预测分数相关用户A，D
    def find_user(self,item_goal):
        user_goal = []
        for key,values in self.data.items():
            for key1,value1 in values.items():
                if(key1 == item_goal):
                    user_goal.append(key)
                    break
        return user_goal

    #2.计算待预测分数中相关用户中两个用户之间的皮尔逊相关系数
    def pearson(self,user1,item_goal): #数据格式为：商品，评分  A:{'a': 4.0, 'c': 3.0, 'd': 5.0
        user_goal = self.find_user(item_goal)
        denominator1 = 0.0 #分母1--待预测用户的分母1
        denominator2 = 0.0 #分母2--相关用户的分母2
        molecule = 0.0     #分子
        r = {}             #皮尔逊系数字典
        try:
            for user2 in user_goal:
                for item_, score_ in self.data[user2].items():
                    for item1, score1 in self.data[user1].items():
                        if(item_ == item1):
                            molecule += (float(score_)-self.ave[user2])*(float(score1)-self.ave[user1])
                            denominator1 += pow(float(score1)-self.ave[user1],2)
                            denominator2 += pow(float(score_)-self.ave[user2],2)
                r.setdefault(user1, {})
                r[user1].setdefault(user2, 0)
                
                # 如果方差为0，直接通过现有的皮尔逊相关系数进行计算
                if(denominator1*denominator2 == 0):
                    # 设定 max 长度
                    max_length = self.max - 1  # 你可以根据需要改变这个值
                    # 创建固定长度的列表，初始化为 0
                    i = [0.0] * max_length
                    j = [0.0] * max_length
                    # 根据键将值放置到指定的位置
                    for key, value in self.data[user1].items():
                        index = int(key) - 1  # 将字符串转换为整数，并减去1以匹配索引
                        if index < max_length:
                            i[index] = value
                    for key, value in self.data[user2].items():
                        index = int(key) - 1  # 同样处理
                        if index < max_length:
                            j[index] = value
                    r[user1][user2] = pearsonr(i, j)[0]
                    
                else:
                    r[user1][user2] = (molecule)/sqrt(denominator1 * denominator2)
                # r[user1][user2] = (molecule)/sqrt(denominator1*denominator2)
                molecule = 0.0
                denominator1 = 0.0
                denominator2 = 0.0
        except Exception as e:
            print("异常信息:", e.message)
            return None
        return r #返回相关用户的皮尔逊系数

    #3.根据皮尔逊系数预测评分
    def prediction(self, user1, item_goal):
        ave1 = self.ave[user1]
        r = self.pearson(user1,item_goal)
        user_goal = self.find_user(item_goal)
        anw1 = 0.0
        anw2 = 0.0
        for user in user_goal:
            anw1 += r[user1][user] * ((self.data[user][item_goal]) - (self.ave[user]))
            anw2 += abs(r[user1][user])
        if(anw2 == 0):
            predict = ave1
        else:
            predict = ave1 + anw1/anw2
        if(predict > 5):
            predict = 5
        self.data[user1][item_goal] = round(predict, 2)

    #扫描数据集，收集未填充数据
    def scan(self):
        item_map = [str(i) for i in range(1,self.max+1)]
        for user,value in self.data.items():
            item_list = []
            for m,n in value.items():
                item_list.append(m)
            for item in item_map:
                if(item not in item_list):
                    self.prediction(user,item)
        for user, value in self.data.items():
            self.data[user] = sorted(self.data[user].items(),key = lambda d:d[0])
        return self.data

if __name__=='__main__':
    users = {
    1: {'1': 5.0, '2': 3.0, '3': 4.0, '4': 3.0, '5': 1.0},
    2: {'1': 3.0, '2': 1.0, '3': 3.0, '4': 3.0, '5': 5.0},
    3: {'1': 4.0, '2': 2.0, '3': 4.0, '4': 1.0, '5': 5.0},
    4: {'1': 4.0, '2': 3.0, '3': 3.0, '4': 5.0, '5': 2.0},
    5: {'2': 3.0, '3': 5.0, '4': 4.0, '5': 1.0}
}
    userCf=UserCf(data=users)
    recommandList=userCf.scan()
    # 转换为 DataFrame
    data = []
    for key, values in recommandList.items():
        for item in values:
            data.append([key, item[0], item[1]])

    # 创建 DataFrame
    df = pd.DataFrame(data, columns=['USER_ID', 'NAME', 'RATING'])
```

```python
import random
import pandas as pd
from math import sqrt
from copy import deepcopy
from sklearn.metrics import mean_squared_error, mean_absolute_error

from scipy.stats import pearsonr

class UserCf():
    def __init__(self, data):
        self.data = data
        self.ave = {}
        self.max = 0
        for key, value in self.data.items():
            sum1 = 0.0
            for item, score in value.items():
                self.max = max(self.max, int(item))
                sum1 += score
            self.ave[key] = sum1 / len(value)

    def find_user(self, item_goal):
        return [key for key, values in self.data.items() if item_goal in values]

    def pearson(self, user1, item_goal):
        user_goal = self.find_user(item_goal)
        r = {}
        for user2 in user_goal:
            molecule = denominator1 = denominator2 = 0.0
            for item in self.data[user1]:
                if item in self.data[user2]:
                    molecule += (self.data[user1][item] - self.ave[user1]) * (self.data[user2][item] - self.ave[user2])
                    denominator1 += (self.data[user1][item] - self.ave[user1]) ** 2
                    denominator2 += (self.data[user2][item] - self.ave[user2]) ** 2
            if denominator1 * denominator2 == 0:
                # 回退皮尔逊
                i = [0.0] * self.max
                j = [0.0] * self.max
                for key, value in self.data[user1].items():
                    i[int(key) - 1] = value
                for key, value in self.data[user2].items():
                    j[int(key) - 1] = value
                corr = pearsonr(i, j)[0]
            else:
                corr = molecule / sqrt(denominator1 * denominator2)
            r.setdefault(user1, {})[user2] = corr
        return r

    def prediction(self, user1, item_goal):
        ave1 = self.ave.get(user1, 0)
        r = self.pearson(user1, item_goal)
        user_goal = self.find_user(item_goal)
        anw1 = anw2 = 0.0
        for user in user_goal:
            anw1 += r[user1][user] * (self.data[user][item_goal] - self.ave[user])
            anw2 += abs(r[user1][user])
        predict = ave1 if anw2 == 0 else ave1 + anw1 / anw2
        return max(0, min(5, round(predict, 2)))  # 限制在[0, 5]

# ========== 🔹构造数据+拆分训练测试集 ==========
def train_test_split(data, test_ratio=0.2, seed=42):
    random.seed(seed)
    train = deepcopy(data)
    test = {}

    for user, items in data.items():
        test[user] = {}
        item_list = list(items.keys())
        test_size = int(len(item_list) * test_ratio)
        test_items = random.sample(item_list, test_size)

        for item in test_items:
            test[user][item] = train[user].pop(item)
    return train, test

# ========== 🔹评估指标 ==========
def evaluate(model, test_data, k=3):
    y_true, y_pred = [], []
    hit, total_pred, total_true = 0, 0, 0
    recommended_items_set = set()

    for user in test_data:
        for item in test_data[user]:
            real = test_data[user][item]
            pred = model.prediction(user, item)
            y_true.append(real)
            y_pred.append(pred)

    mae = mean_absolute_error(y_true, y_pred)
    rmse = sqrt(mean_squared_error(y_true, y_pred))

    for user in test_data:
        scored_items = {}
        for i in range(1, model.max + 1):
            item = str(i)
            if item not in model.data[user]:
                pred = model.prediction(user, item)
                scored_items[item] = pred
        top_k = sorted(scored_items.items(), key=lambda x: x[1], reverse=True)[:k]
        true_items = set(test_data[user].keys())
        rec_items = set([item for item, score in top_k])
        
        recommended_items_set.update(rec_items)
        hit += len(true_items & rec_items)
        total_pred += len(rec_items)
        total_true += len(true_items)

    precision = hit / total_pred if total_pred else 0
    recall = hit / total_true if total_true else 0
    f1 = (2 * precision * recall) / (precision + recall) if (precision + recall) else 0

    all_items = set(str(i) for i in range(1, model.max + 1))
    coverage = len(recommended_items_set) / len(all_items)

    return {
        "MAE": round(mae, 4),
        "RMSE": round(rmse, 4),
        "Precision@K": round(precision, 4),
        "Recall@K": round(recall, 4),
        "F1@K": round(f1, 4),
        "Coverage": round(coverage, 4)
    }


# ========== 🔹运行 ==========
if __name__ == '__main__':
    users = {
        1: {'1': 5.0, '2': 3.0, '3': 4.0, '4': 3.0, '5': 1.0},
        2: {'1': 3.0, '2': 1.0, '3': 3.0, '4': 3.0, '5': 5.0},
        3: {'1': 4.0, '2': 2.0, '3': 4.0, '4': 1.0, '5': 5.0},
        4: {'1': 4.0, '2': 3.0, '3': 3.0, '4': 5.0, '5': 2.0, '6': 5.0, '7': 3.0},
        5: {'2': 3.0, '3': 5.0, '4': 4.0, '5': 1.0, '6': 1.0}
    }

    train_data, test_data = train_test_split(users, test_ratio=0.4)
    model = UserCf(train_data)

    results = evaluate(model, test_data)
    print("不同 K 值下的推荐系统评估指标：\n")
    for k in [3, 5, 10, 15, 20]:
        results = evaluate(model, test_data, k=k)
        print(f"🔹 K = {k}")
        for key, value in results.items():
            print(f"  {key}: {value}")
        print()
```

```python
num = len(df_filter)
def cal_precision(df):
    # 计算 pre_re 和 preference 都为 1 的数量
    count_both_one = len(df[(df['pre_re'] == 1) & (df['preference'] == 1)])

    # 计算 pre_re 为 1 的数量
    count_pre_re_one = len(df[df['pre_re'] == 1])
    

    # 计算比例
    if count_pre_re_one > 0:
        ratio = count_both_one / count_pre_re_one
    else:
        ratio = 0  # 避免除以0的情况

    return ratio

def cal_recall(df):
    # 计算 pre_re 和 preference 都为 1 的数量
    count_both_one = len(df[(df['pre_re'] == 1) & (df['preference'] == 1)])

    count_pre_re_one = num

    # 计算比例
    if count_pre_re_one > 0:
        ratio = count_both_one / count_pre_re_one
    else:
        ratio = 0  # 避免除以0的情况

    return ratio

print("基于用户的协同过滤准确率：", cal_precision(df_user), "召回率：", cal_recall(df_user))
print("基于项目的协同过滤准确率：", cal_precision(df_item), "召回率：", cal_recall(df_item))
```
